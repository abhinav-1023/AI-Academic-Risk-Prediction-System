Advanced Synthetic Dataset Code

```python
import pandas as pd
import numpy as np

# For reproducibility
np.random.seed(42)

# Number of students
n = 800

# Generate base features
data = pd.DataFrame({
    "attendance": np.random.randint(50, 100, n),
    "study_hours": np.random.randint(1, 35, n),
    "assignments_avg": np.random.randint(40, 100, n),
    "internal_marks": np.random.randint(30, 100, n),
    "previous_gpa": np.round(np.random.uniform(4, 10, n), 2),
    "absences": np.random.randint(0, 25, n),
    "participation_score": np.random.randint(1, 10, n),
    "internet_usage_hours": np.random.randint(1, 8, n),
    "sleep_hours": np.random.randint(4, 9, n)
})

# Add small realistic noise to marks
data["internal_marks"] = data["internal_marks"] + np.random.normal(0, 5, n)
data["assignments_avg"] = data["assignments_avg"] + np.random.normal(0, 5, n)

# Keep values within valid range
data["internal_marks"] = data["internal_marks"].clip(0, 100)
data["assignments_avg"] = data["assignments_avg"].clip(0, 100)

# Risk classification logic (3 classes)
conditions = [
    (data["attendance"] < 60) |
    (data["internal_marks"] < 40) |
    (data["study_hours"] < 5),

    (data["attendance"] < 75) |
    (data["internal_marks"] < 60),

    (data["attendance"] >= 75) &
    (data["internal_marks"] >= 60)
]

choices = ["High", "Medium", "Low"]

data["risk_level"] = np.select(conditions, choices, default="Medium")

# Add 5% random label noise (to make it realistic)
mask = np.random.rand(n) < 0.05
data.loc[mask, "risk_level"] = np.random.choice(
    ["Low", "Medium", "High"], 
    mask.sum()
)

# Save dataset
data.to_csv("../data/student_data.csv", index=False)

# Show first few rows
data.head()
```
